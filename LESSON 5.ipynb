```python
#STRING INDEXING  ~ Python starts from zero

word = "PYTHON"

#Positive indexing (left to right)
print(word[0]) # First letter
print(word[1]) #Second letter

#Negative indexing (Right to left)
print(word[-1]) #last letter
```

```python
#STRING SLICING ~ the syntax is [start:stop] NOTE the stop index is exlusive(not included in the result)
text = "NAIROBI"

#Grab from index 0 upto (but not including ) 3
print(text[0:3])

#Grab from index 3 to the end 
print(text[3:])
```

```python
##OPERATIONS IN STRINGS
#STRING CONCATENATION ~ To join or repeat  
```

```python
#Concatenation (+)
a = "HARRY"
b = "MAGUIRE"
print(a + " " + b)

#Repetition (*)
word = "Goal!"
print("Goal! " *3)
```

```python
#ESCAPE CHARACTERS
# \n ~ New line (moves to the next line)
# \t ~ Tab (add a large space)
#\' or \" ~ Quotes (print a quote sysmbol
#\\  ~ Backslash (prints a literal backslash) 

print("My name is Tsungi.\nl am 23 yers old.")
print('It\'s a beautiful day')
```

```python
##STRING REPETITION

print("HARRY"*4) #Repeats without spacing
print("HARRY "*4) #Repeats with spacing

#check a string type
print(type ("HARRY"))
```

# VARIABLES

```python
#Variables ~ name or label refers to a value stored in the computer's memory. Allows to store or retrieve data and manipulate it later

#The must Dos ~ Variables start with a letter (a-z,A-Z) or an underscore(_)
#The must Dos ~ Variables contain only letters, numbers, or underscores
#The must Dos ~ Variables are case sensitive : age, Age, AGE are three different things

first_name = "Tsungi"
age_2024 = 23
_secrete_code = 1234

#The Don'ts ~ No starting with a number (eg 1st_prize is invalid)
#The Don'ts ~ No spaces (eg my name is invalid) Use underscores instead (my_name)
#The Don'ts ~ No using python keywords (words tha mean speacial to python like print,if,for, True) print = "text" is invalid 


```

```python
#Naming Conventions (Snake case) means using all lowercase letters and separating words with underscores)
first_name = "Harry"
second_name = "Maguire"

print(first_name)
print(second_name)
```

```python
#Re-assingment ~ Variables can vary(change)

score = 0
print("Start of the match:", score)

score = 1
print("After first goal:", score)

score = score +1
print("After second goal:", score) 
```

```python
x = 10
x = 10+1

print (x)

x = 10
x += 1

print(x)
```

```python
#VARIABLE  MULTIPLE ASSIGNMENTS
#Assigning different values to different variables

mike, joyce, vic = 23, 56, 11

print(joyce)
print(joyce,mike)

#Assigning the SAME value to multiple variables

a = b = c = 100
print(a)
print(b) 
```
