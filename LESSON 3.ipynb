# OPERATORS

```python
#ADDITION (+)

print(21+76)
```

```python
#SUBTRACTION (-)

print(80-10)
```

```python
#MULTIPLICATION (*)

print(7*9)
```

```python
#DIVISION (/) ~ standard division always returns a Float (a number with a decimal point), even if the division is clean.

print(20/10)
```

```python
#FLOOR DIVISION (//) ~ divides numbers but chops off the decimal part, leaving only the whole integer

print(20//10)
print(20/10.0)
```

```python
# MODULUS

print(10 % 3 )
```

```python
#EXPONENT (**)

print(2 ** 3)
```

```python
#PRECEDENCE AND ASSOCIATIVITY
# Multiplication(*) beats Addition (+)

10 + 5 * 2
```

```python
#Parentheses () beats everything

(10+5) * 2 
```

```python
#Exponentiation is Right-to-left
# 2 ** 3 ** 2 means 2 ** (3 ** 2) = 2 ** 9

2 ** 3 ** 2
```

```python
print(11 // 2 + 11 % 2 )

print((5 + 5) * 2**2)

print(10 + 2 % 3)

(100 / 10 * 10)
```
